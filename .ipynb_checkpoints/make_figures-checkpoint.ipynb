# __Helper Functions__

```python
def round_to_50(x, base=50):
    # Takes in a number a rounds it to the nearest 50 up in magnitude
    # Used for bounds in S6
    if x >= 0: #for positive #s, round up
        return base * np.ceil(x/base)
    else: #for negative #s, round down
        return base * np.floor(x/base)
```

```python
def get_attr(rout, attr='bf'):
    # Takes in an rpy2 vector, returns the value of an attribute
    # Used for BayesFactor in 2c
        try:
            index = list(rout.names).index(attr)
            val = list(rout.items())[index][1]
            if len(val) == 1:
                val = val[0]
            if type(val)==rpy2.robjects.vectors.Matrix:
                val = np.asarray(val)
            return val
        except ValueError:
            print('Did not pass a valid attribute')
```

```python
def calc_SSRT_wOmission(subj_df):

    #Set up keys for stop trial types
    stopTrialKey = 'stop'
    if 'Stop' in subj_df['TrialType'].unique(): stopTrialKey = 'Stop' #ADDED FOR SACCADE TASK

    #Set up keys for go trial types
    goTrialKey = 'go'
    goRT_key = 'GoRT'

    if 'GoCritical' in subj_df['TrialType'].unique(): #ADDED FOR TURK MOTOR SELECTIVE STOP
        goTrialKey = 'GoCritical'
        goRT_key = 'GoCriticalRT'


    ################################################
    # Getting MaxRT, P(respond|Signal), mean_SSD #, omission #s

    maxRT = np.maximum(subj_df.loc[:, goRT_key].max(),subj_df['StopFailureRT'].max())  #get the max RT to replace go omissions


    # Calculate P(respond | signal) as num_stop_failures / num_stop_trials
    num_stop_failures = subj_df.loc[(subj_df['TrialType']==stopTrialKey) & (subj_df['StopFailureRT']>0), 'StopFailureRT'].count()
    num_stop_trials = len(subj_df.loc[subj_df['TrialType']==stopTrialKey, 'StopFailureRT'])
    P_respond = num_stop_failures / num_stop_trials

    #get the mean SSD to subtract from the nth RT
    mean_SSD = subj_df['StopSignalDelay'].mean()

    #get the omission count and the omission rate
    goRTs = subj_df.loc[(subj_df['TrialType']==goTrialKey) & (subj_df[goRT_key]>0), goRT_key].values
    goRTs.sort()
    num_go_trials = subj_df.loc[subj_df['TrialType']==goTrialKey, 'TrialType'].count()
    num_go_responses = len(goRTs)

    omission_count = num_go_trials - num_go_responses
    omission_rate = omission_count/num_go_trials

    #Correct P(respond | signal)
    corrected_P_respond = P_respond/(1-omission_rate)

    #get nth RT using corrected P(respond | signal)
    nth_index = int(np.ceil(corrected_P_respond*num_go_responses)) - 1
    if nth_index >= len(goRTs): #if nth_index too large, take largest RT
        nth_RT = goRTs[-1]
    else:
        nth_RT = goRTs[nth_index]

    
    
    

    ################################################
    # Calculate SSRT as nth_RT - mean_SSD
    SSRT = nth_RT - mean_SSD

    return SSRT, mean_SSD
```

```python
def calc_SSRT_wReplacement(subj_df):
    
    #Set up keys for stop trial types
    stopTrialKey = 'stop'
    if 'Stop' in subj_df['TrialType'].unique(): stopTrialKey = 'Stop' #ADDED FOR SACCADE TASK

    #Set up keys for go trial types
    goTrialKeys = ['go']
    goRT_keys = ['GoRT']
    
    if 'GoCritical' in subj_df['TrialType'].unique(): #ADDED FOR TURK MOTOR SELECTIVE STOP
        goTrialKeys = ['GoCritical']
        goRT_keys = ['GoCriticalRT']
    

    ################################################
    # Getting MaxRT, P(respond|Signal), mean_SSD #

    maxRT = np.maximum(subj_df.loc[:, goRT_keys].max().max(),subj_df['StopFailureRT'].max())  #get the max RT to replace go omissions

    # Calculate P(respond | signal) as num_stop_failures / num_stop_trials
    num_stop_failures = subj_df.loc[(subj_df['TrialType']==stopTrialKey) & (subj_df['StopFailureRT']>0), 'StopFailureRT'].count()
    num_stop_trials = len(subj_df.loc[subj_df['TrialType']==stopTrialKey, 'StopFailureRT'])
    P_respond = num_stop_failures / num_stop_trials

    #get the mean SSD to subtract from the nth RT
    mean_SSD = subj_df['StopSignalDelay'].mean()

    ################################################
    ############## Getting Nth RT ##################


    # 1 - create index of RTs to include 
    subj_df['include_for_SSRT'] = 0

    subj_df.loc[subj_df['TrialType'].isin(goTrialKeys), 'include_for_SSRT'] = 1 #add all go trials 
    subj_df.loc[(subj_df['StopFailureRT']<subj_df['StopSignalDelay'])\
                & (subj_df['StopFailureRT']>0.0), 'include_for_SSRT'] = 1 #add all stopFailures greater than 0 but faster than SSD

    # 2 - get values for each RT being included
    subj_df['RTs_for_SSRT'] = np.nan
    # add in all GoRTs
    goRT_df = subj_df.loc[subj_df['include_for_SSRT']==1, goRT_keys].copy() #subset to the goRT columns
    goRTs = goRT_df.max(axis=1) # get the nonNaN from each row
    subj_df.loc[subj_df['include_for_SSRT']==1, 'RTs_for_SSRT'] = goRTs 

    # fill go ommissions (and fast stopFailureRTs) with max GoRt
    if len(subj_df.loc[(subj_df['include_for_SSRT']==1) & (subj_df['RTs_for_SSRT'].isnull())]) > 0:
        subj_df.loc[(subj_df['include_for_SSRT']==1) & (subj_df['RTs_for_SSRT'].isnull()), 'RTs_for_SSRT'] = maxRT

    # overwrite StopFailureRTs with actual values
    subj_df.loc[(subj_df['include_for_SSRT']==1) & subj_df['StopFailureRT']>0.0, 'RTs_for_SSRT'] = \
    subj_df.loc[(subj_df['include_for_SSRT']==1) & subj_df['StopFailureRT']>0.0, 'StopFailureRT']

    # 3 - get Nth RT
    # get RTs used for SSRT and sort 
    useful_RTs = subj_df['RTs_for_SSRT'].values.copy()
    useful_RTs = [i for i in useful_RTs if i==i]
    useful_RTs.sort()
    #Get the number of RTs to be used
    num_RTs = len(useful_RTs)

    # Get the nth_RT based on P(respond | Signal)
    nth_index = int(np.ceil(P_respond*num_RTs)) - 1
    if nth_index >= len(useful_RTs): #if nth_index too large, take largest RT
        nth_RT = useful_RTs[-1]
    else:
        nth_RT = useful_RTs[nth_index]
    ################################################

    # Calculate SSRT as nth_RT - mean_SSD
    SSRT = nth_RT - mean_SSD

    return SSRT, mean_SSD
```

```python
def calc_SSRT(subj_df, method='omission'):
    if method=='omission':
        return calc_SSRT_wOmission(subj_df)
    elif method=='replacement':
        return calc_SSRT_wReplacement(subj_df)
    else:
        raise Exception("SSRT method must be either 'omission' or 'replacement'")
```

```python
cumulativeStopTrials_per_ssd = None
for data_file in data_files:
    condition_name = data_file.replace('Data/Data', '').replace('.xlsx', '') #get condition
    curr_df = pd.read_excel(data_file)
    nstoptrials_per_ssd = curr_df.groupby('StopSignalDelay').count()['Subject'] #sum num stop trials across subjects at each SSD
    nstoptrials_per_ssd = nstoptrials_per_ssd.drop(labels=['?']) #drop non stop trials
    cumulative_percentages_per_ssd = nstoptrials_per_ssd.cumsum() / nstoptrials_per_ssd.sum() #take the cumulativesum / total
    cumulative_percentages_per_ssd.name = condition_name #name the series after the condition
    if cumulativeStopTrials_per_ssd is None: #create or add the series to a full DF
        cumulativeStopTrials_per_ssd = cumulative_percentages_per_ssd
    else:
        cumulativeStopTrials_per_ssd = pd.concat([cumulativeStopTrials_per_ssd,
                                                   cumulative_percentages_per_ssd], axis=1, sort=True)
cumulativeStopTrials_per_ssd = cumulativeStopTrials_per_ssd[short_names]  
```

```python
plt.figure(figsize=(12,8))
plt.rc('axes', prop_cycle=(cycler('color', linecolors) +
                           cycler('linestyle', linestyles)))

# plot the data
plt.plot(cumulativeStopTrials_per_ssd)
plt.axis([0,750,0,1])
# plot mean
plt.plot(cumulativeStopTrials_per_ssd.mean(1),linewidth=4,color='k')
ax=plt.gca()

plt.xlabel('Stop-signal delay (ms)',fontsize=24)
plt.ylabel('Cumulative Proportion of SSDs',fontsize=24)
xticks=np.arange(0,800,50)
ax.set_xticks(xticks)
ax.set_xticklabels(xticks,fontsize=18)
yticks=np.arange(0,1.01,.1)
ax.set_yticks(yticks)
ax.set_yticklabels(['%0.1f'%i for i in yticks],fontsize=18)

plt.savefig(path.join(FIGURE_DIR, 'figure1d.png'))
plt.show()
_ = plt.close()
```

```python
# Load in data
cond20_file = 'Data/DataBtwnSubjStimSelec.xlsx'
cond20_df = pd.read_excel(cond20_file)
cond20_df = cond20_df.replace(r'^\s*$', np.nan, regex=True).replace('?', np.nan)
```

```python
####################################################################################################
# For each subject, classify them as SD or DDS
# Using bayesfactor as done in the original paper (which used http://pcl.missouri.edu/bf-two-sample)

strategy_labels = []

for subject in cond20_df.Subject.unique():
    subdata = cond20_df.query('Subject == %d' % subject).copy()
    # Get distribution of goRTs, signalRespondRTs, and ignoreRTs for t-tests and subsequent BayeFactor
    goRTs = subdata.loc[(subdata['TrialType']=='go') & (subdata['GoRT']>0), 'GoRT'].values.copy()
    signalRespondRTs = subdata.loc[(subdata['TrialType']=='stop') & (subdata['StopFailureRT']>0), 'StopFailureRT'].values.copy()
    ignoreRTs = subdata.loc[(subdata['TrialType']=='ignore') & (subdata['IgnoreRT']>0), 'IgnoreRT'].values.copy()
    t_signalRespond_v_noSignal, _ = stats.ttest_ind(signalRespondRTs, goRTs)
    t_ignore_v_noSignal, _ = stats.ttest_ind(ignoreRTs, goRTs)
    bayes_out_signalRespond = BayesFactor.ttest_tstat(t=float(t_signalRespond_v_noSignal),
                                                      n1=len(signalRespondRTs),
                                                      n2=len(goRTs), rscale = 1.0) #using 1 as in the 2014 paper
    bf_signalRespond = math.exp(get_attr(bayes_out_signalRespond))
    
    bayes_out_ignore = BayesFactor.ttest_tstat(t=float(t_ignore_v_noSignal),
                                                      n1=len(ignoreRTs),
                                                      n2=len(goRTs), rscale = 1.0) #using 1 as in the 2014 paper
    bf_ignore = math.exp(get_attr(bayes_out_ignore))
    
    if np.abs(bf_ignore) > 1: #for both groups, must favor the alternative for ignore vs no-signal
        if np.abs(bf_signalRespond) > 1: 
            strategy_labels.append('SD')
        else:
            strategy_labels.append('DDS')
    else:
        strategy_labels.labels('uncategorized/IDS')
```

```python
###################################################################################################
# Generate a DF that has every stop trial which had a previous Go trial, contains the previous goRT
verbose = False
go_key = 'GoRT'


stop_trials_w_matchingGo = None
for subject in cond20_df.Subject.unique():
    subdata = cond20_df.query('Subject == %d' % subject).copy()
    if verbose: print('subject %d: found %d trials' % (subject, subdata.shape[0]))
    ssdvals = [i for i in subdata.StopSignalDelay.unique() if isinstance(i, float) and i >=0] #ignore nan/missing values and 0/negative SSDs
    ssdvals.sort()

    # You find all pairs of trials in which the first is a go trials with a response and the second is a 
    # stop trial with a response. Both trials need to come from the same subject and block. 
    # This should be done separately for each SSD. Then the RT for the first trial in the pair should be 
    # subtracted from the second. This is the core analysis per SSD per subject.

    for ssd in ssdvals:
        ssd_data = subdata.query('StopSignalDelay == %d' % ssd).copy()
        signal_respond_data = ssd_data.dropna(subset=["StopFailureRT"])
        signal_respond_data = signal_respond_data.loc[signal_respond_data['StopFailureRT'] > 0, :]
        # for each signal respond trial, determine whether the previous trial was a go trial in the same block
        signal_respond_data['MatchingGo'] = np.nan
        if signal_respond_data.shape[0] > 0:
            for t in signal_respond_data.index:
                if t < 1:
                    continue
                prevtrial = cond20_df.loc[int(t) - 1, :]
                if prevtrial.Block == signal_respond_data.loc[t, 'Block']:
                    # blockmatch
                    signal_respond_data.loc[t, 'MatchingGo'] = prevtrial[go_key]
            signal_respond_data = signal_respond_data.dropna(subset=['MatchingGo'])
            if stop_trials_w_matchingGo is None:
                stop_trials_w_matchingGo = signal_respond_data
            else:
                stop_trials_w_matchingGo = pd.concat([stop_trials_w_matchingGo, signal_respond_data], axis=0)
```

```python
#Build up pertinent data for each subject: StopFailRTAtShorterSSDs, StopFailRTAtLongerSSDs, PrecedingNoStopRTAtShorterSSDs, PrecedingNoStopRTAtLongerSSDs
stopFailRTs_shortSSDs = []
stopFailRTs_longSSDs = []
precedingNoStopRTs_shortSSDs = []
precedingNoStopRTs_longSSDs = []

for subject in stop_trials_w_matchingGo.Subject.unique():
    subdata = stop_trials_w_matchingGo.query('Subject == %d' % subject).copy()
    
    shortSSD_df = subdata.loc[subdata['StopSignalDelay']<250]
    stopFailRTs_shortSSDs.append(shortSSD_df['StopFailureRT'].mean())
    precedingNoStopRTs_shortSSDs.append(shortSSD_df['MatchingGo'].mean())
    
    longSSD_df = subdata.loc[subdata['StopSignalDelay']<250]
    stopFailRTs_longSSDs.append(longSSD_df['StopFailureRT'].mean())
    precedingNoStopRTs_longSSDs.append(longSSD_df['MatchingGo'].mean())
```

```python
#Combine into a single df
fig2c_df = pd.DataFrame({'Strategy': strategy_labels, 'StopFailRTAtShorterSSDs': stopFailRTs_shortSSDs,
                        'StopFailRTAtLongerSSDs': stopFailRTs_longSSDs, 'PrecedingNoStopRTAtShorterSSDs': precedingNoStopRTs_shortSSDs, 
                        'PrecedingNoStopRTAtLongerSSDs': precedingNoStopRTs_longSSDs})
fig2c_df = fig2c_df.sort_values(by=['Strategy'], ascending=False)
fig2c_df = fig2c_df.dropna() # 3 subjects were identified as having nans for each value!!

data_2c = fig2c_df.replace({'below':'Mean SSD < 250 ms',
                                    'above':'Mean SSD > 250 ms'})

data_2c_tidy=fig2c_df.melt(id_vars='Strategy',
                            value_vars=['StopFailRTAtShorterSSDs', 
                                        'StopFailRTAtLongerSSDs',
                                        'PrecedingNoStopRTAtShorterSSDs', 
                                        'PrecedingNoStopRTAtLongerSSDs'],
                            var_name='SSD', 
                            value_name='Reaction time (ms)')
data_2c_tidy['Type']=data_2c_tidy['SSD']
data_2c_tidy['ShortLong']=data_2c_tidy['SSD']


data_2c_tidy.loc[data_2c_tidy['SSD']=='PrecedingNoStopRTAtLongerSSDs','ShortLong']='Mean SSD > 250 ms'
data_2c_tidy.loc[data_2c_tidy['SSD']=='StopFailRTAtLongerSSDs','ShortLong']='Mean SSD > 250 ms'
data_2c_tidy.loc[data_2c_tidy['SSD']=='PrecedingNoStopRTAtShorterSSDs','ShortLong']='Mean SSD < 250 ms'
data_2c_tidy.loc[data_2c_tidy['SSD']=='StopFailRTAtShorterSSDs','ShortLong']='Mean SSD < 250 ms'

data_2c_tidy.loc[data_2c_tidy['SSD']=='PrecedingNoStopRTAtLongerSSDs','Type']='PrecedingNoStop'
data_2c_tidy.loc[data_2c_tidy['SSD']=='StopFailRTAtLongerSSDs','Type']='StopFail'
data_2c_tidy.loc[data_2c_tidy['SSD']=='PrecedingNoStopRTAtShorterSSDs','Type']='PrecedingNoStop'
data_2c_tidy.loc[data_2c_tidy['SSD']=='StopFailRTAtShorterSSDs','Type']='StopFail'


del data_2c_tidy['SSD']


ViolationShortSSD=fig2c_df['PrecedingNoStopRTAtShorterSSDs']-fig2c_df['StopFailRTAtShorterSSDs']
ViolationLongSSD=fig2c_df['PrecedingNoStopRTAtLongerSSDs']-fig2c_df['StopFailRTAtLongerSSDs']

data_2c_violation=fig2c_df.assign(ViolationShortSSD=ViolationShortSSD,
                                ViolationLongSSD=ViolationLongSSD)
for i in ['StopFailRTAtShorterSSDs', 'StopFailRTAtLongerSSDs',
       'PrecedingNoStopRTAtShorterSSDs', 'PrecedingNoStopRTAtLongerSSDs']:
    del data_2c_violation[i]
    
data_2c_violation_tidy=data_2c_violation.melt(id_vars='Strategy',
                            value_vars=['ViolationShortSSD','ViolationLongSSD'],
                            var_name='SSD', 
                            value_name='No-stop RT - Stop failure RT')
```

```python
# for each condition, get a mean SSRT using all SSDs, and only using SSDs >= 200ms
condition_order = []
SSRTs_per_condition = []
SSRTs_longSSDs_per_condition = []

for data_file in data_files:
    cond_SSRTs = []
    cond_SSRTs_longSSDs = []
    condition_name = data_file.replace('Data/Data', '').replace('.xlsx', '') #get condition
    curr_df = pd.read_excel(data_file)
    curr_df = curr_df.replace(r'^\s*$', np.nan, regex=True).replace('?', np.nan)
    
    condition_subjects = curr_df.Subject.unique()
    if 'Turk' in condition_name: condition_subjects = passed_turkers
    for subject in condition_subjects:
        #First get SSRT using all SSDs (after dropping negative SSDs)   
        subdata = curr_df.query('Subject == %d' % subject).copy()
        #subset to ignore neg SSDs, keep NaNs( = go and ignore rows) - NO LONGER
#         subdata = subdata[(subdata.StopSignalDelay >= 0) | (subdata.StopSignalDelay.isnull())]
        SSRT, _ = calc_SSRT(subdata, method='replacement')
        cond_SSRTs.append(SSRT)
        
        #Get SSRT after excluding SSDs < 200ms
        if len(subdata[(subdata.StopSignalDelay >= 200)]) > 0: #Some subjects have no SSDs >= 200
            subdata = subdata[(subdata.StopSignalDelay >= 200) | (subdata.StopSignalDelay.isnull())]
            SSRT, _ = calc_SSRT(subdata, method='replacement')
            cond_SSRTs_longSSDs.append(SSRT)
    
    condition_order.append(condition_name)
    SSRTs_per_condition.append(np.asarray(cond_SSRTs))
    SSRTs_longSSDs_per_condition.append(np.asarray(cond_SSRTs_longSSDs))

# Get mean SSRT (full and long SSDs) for each condition 
meanSSRTs_per_condition = [np.nanmean(cond_SSRTs) for cond_SSRTs in list(SSRTs_per_condition)]
meanSSRTs_longSSDs_per_condition = [np.nanmean(cond_SSRTs) for cond_SSRTs in list(SSRTs_longSSDs_per_condition)]


#build up dictionary - k,v = condition_name, [full SSDs SSRT, long SSDs SSRT] - to be converted into df
data2d_dict = {}
for idx, condition in enumerate(condition_order):
    data2d_dict[condition] = [meanSSRTs_per_condition[idx], meanSSRTs_longSSDs_per_condition[idx]]

# transform into DF
data2d_df = pd.DataFrame(data2d_dict)
data2d_df = data2d_df[short_names] # reorder for consistent order across figures

#PLOT
plt.figure(figsize=(9,8))
plt.rc('axes', prop_cycle=(cycler('color', linecolors) +
                           cycler('linestyle', linestyles)))

# plot the data
plt.plot(data2d_df)

plt.axis([-0.25,1.25, 0, 450]) # previously [.75,2.25, 0,350]

# plot mean
plt.plot(data2d_df.mean(1),linewidth=4,color='k')

# Add the patch to the Axes
ax=plt.gca()

# plt.xlabel('All SSDs                           SSDs >= 200ms',fontsize=24)
plt.ylabel('SSRT (ms)',fontsize=24)
ax.set_xticks([0, 1])
ax.set_xticklabels(['All SSDs', 'SSDs >= 200ms'], fontsize=24)

yticks=np.arange(0,500,50) #previously np.arange(0,400,50)
ax.set_yticks(yticks)
ax.set_yticklabels(yticks,fontsize=18)

plt.savefig(path.join(FIGURE_DIR, 'figure2d.png'), bbox_inches='tight')
plt.show()
_ = plt.close()
```

```python
SSRTs_per_ssd_per_subj = []

for subject in fixedSSD2_df.Subject.unique():
    #First get SSRT using all SSDs (after dropping negative SSDs)   
    subdata = fixedSSD2_df.query('Subject == %d' % subject).copy()
    ssdvals = [i for i in subdata.StopSignalDelay.unique() if i==i and i >=0] #ignore nan/missing values and 0/negative SSD
    
    SSRTs_per_ssd = []
    for ssdval in ssdvals:
        #Get SSRT using only stop trials at that SSD
        if len(subdata[(subdata.StopSignalDelay == ssdval)]) > 0: #make sure they have some data at the ssdval
            ssd_data = subdata[(subdata.StopSignalDelay == ssdval) | (subdata.StopSignalDelay.isnull())].copy()
            SSRT, _ = calc_SSRT(ssd_data)
            SSRTs_per_ssd.append(SSRT)
        else:
            SSRTs_per_ssd.append(np.nan)
    SSRTs_per_ssd_per_subj.append(SSRTs_per_ssd)
    
#Make into df    
SSRTs_per_ssd_per_subj = np.asarray(SSRTs_per_ssd_per_subj).T
ssd_range = np.arange(0, 550, 50)
SSRTs_per_ssd_per_subj = pd.DataFrame(SSRTs_per_ssd_per_subj, columns=np.arange(1, len(fixedSSD2_df.Subject.unique())+1), index=ssd_range)
```
